## 1. Calculate the filter in each layer


### CONV Layer:

**Number of parameters = ((shape of width of the filter * shape of height of the filter * number of filters in the previous layer + 1) * number of filters in current layer)**

Apply for **conv2d (None, 28, 28, 20)**, **param = 520**:

* Filter: k x k

* Number of filters in the previous layer: 1

* Number of filters in current layer: 20

Calculate:
520 = ((k * k * 1) + 1) * 20 => k = 5

*Because output size is same to input size, stride of CONV layer s = 1.*

**Shape of output size = (shape of input size - shape of filter + 2 * zero padding)/stride + 1**

=> zero padding p = 2.

**Conclusion: CONV1(Filter: 5x5, Stride: 1, Padding: 2)**

### POOL Layer

**Number of parameters = 0**

This has got no learnable parameters because all it does is calculate a specific number, no backprop learning involved.

### Fully Connected Layer (FC)

**Number of parameters = ((current layer neurons * previous layer neurons) + 1 * current layer neurons).**

* Previous layer neurons: 20 * 14 * 14
* Current layer neurons: 10


## 2. Code in Pytorch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchsummary import summary

class CNNs(nn.Module):  # Káº¿ thá»«a tá»« nn.Module
    def __init__(self, num_channels = 1):
        super(CNNs, self).__init__()
        # Conv2D, MaxPooling and Fully Connected layers
        self.conv2d = nn.Conv2d(in_channels = num_channels, out_channels = 20, kernel_size = (5, 5), stride = 1, padding = 2)
        self.max_pooling2d = nn.MaxPool2d(kernel_size = (2,2), stride = (2, 2))
        self.flatten = nn.Flatten()
        self.dense = nn.Linear(in_features = 14*14*20, out_features = 10)

    def forward(self, x):
        x = x.unsqueeze(1)
        x = F.relu(self.conv2d(x))
        x = self.max_pooling2d(x)
        x = self.flatten(x)
        x = self.dense(x)
        return x
```

```python
model = CNNs(num_channels=1)
summary(model, input_size=(28, 28))
```
